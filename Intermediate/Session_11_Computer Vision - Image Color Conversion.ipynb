```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
#Import image
file_name = "dog.png"
image = cv2.imread(file_name, cv2.IMREAD_COLOR)
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

print(rgb_image.shape)
print(rgb_image[800][700])

#Show the image with matplotlib
plt.imshow(rgb_image)
plt.show()
```

```python
#Import image
image = cv2.imread(file_name, cv2.IMREAD_COLOR)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
print(gray_image.shape)
#Show the image with matplotlib
plt.imshow(gray_image, cmap='gray')
plt.show()
```

```python
#Import image
image = cv2.imread(file_name, cv2.IMREAD_COLOR)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

thresh = 2
bw_image = cv2.threshold(gray_image, thresh, 255, cv2.THRESH_BINARY)[1]
plt.imshow(bw_image, cmap='Greys', interpolation='nearest')
plt.show()
```

```python
invert_image = cv2.bitwise_not(bw_image)
plt.imshow(invert_image, cmap='Greys', interpolation='nearest')
plt.show()
```
